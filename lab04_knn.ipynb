```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()
X = iris.data
y = iris.target


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)


model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

print("Enter Flower Details:")
sepal_length = float(input("Sepal Length: "))
sepal_width = float(input("Sepal Width: "))
petal_length = float(input("Petal Length: "))
petal_width = float(input("Petal Width: "))

user_input = [[sepal_length, sepal_width, petal_length, petal_width]]


prediction = model.predict(user_input)

print("\nPredicted Flower Type:", iris.target_names[prediction][0])
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler

diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)


model = KNeighborsRegressor(n_neighbors=5)
model.fit(X_train, y_train)

print("Enter REALISTIC patient details:\n")

age = float(input("Age (years): "))
sex = float(input("Sex (0=Female, 1=Male): "))
bmi = float(input("BMI: "))
bp = float(input("Blood Pressure: "))
tc = float(input("Total Cholesterol: "))
ldl = float(input("LDL: "))
hdl = float(input("HDL: "))
tch = float(input("TCH: "))
ltg = float(input("LTG: "))
glu = float(input("Glucose: "))

user_input = [[age, sex, bmi, bp, tc, ldl, hdl, tch, ltg, glu]]


user_input_scaled = scaler.transform(user_input)


prediction = model.predict(user_input_scaled)

print("\nPredicted Disease Progression Value:", prediction[0])
```
